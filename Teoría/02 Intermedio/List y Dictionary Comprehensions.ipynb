# List Comprehensions
##
---
##
#### Creación de listas de forma simple con el uso de una sola línea de código.
##
---
##
###### La nomenclatura a utilizar para establecer una list comprehension es llenar dentro de []:
###### 1. La fórmula iterable para definir el contenido de cada campo
###### 2. El ciclo for en la que se moverá la variable iterable
###### 3. La condición, en caso se requiera, para filtrar los datos que no deben ingresar al arreglo
##
```
# Para este ejemplo, se quiere crear una lista de número del 0 al 4:
lst = [i for i in range(5)]
```
##
```
# Para este ejemplo, se crea una lista del doble de aquellos números pares del 1 al 5:
lst = [i*2 for i in range(1,5) if i%2==0]
```

# Dictionary Comprehensions
---
## 
#### De igual forma, se pueden crear diccionarios de forma simple con una sola línea de código
## 
---
##
###### Se utiliza la siguiente fórmula que sigue la misma lógica que el de listas, solo que ahora debemos indicar las llaves y valores al mismo tiempo:
##
```
dct = {a:b*2 for a in range(1,10) if a%2==0}
```
###### Nota: En este ejemplo, se lista como llave los números pares del 1 al 100, y cuyos valores son su llave multiplicada por 2. 

# Ejemplos de List Comprehensions:

````
# Listado del 0 al 9:
l = [i for i in range(10)]
print(l)

# Números pares del 1 al 5 multiplicados por 2:
lst = [i*2 for i in range(1,5) if i%2==0]
print(lst)
````

# Ejemplos de Dictionary Comprehensions:

````
# Llaves: Los 100 primeros números; Valores: Su llave elevada al cubo
dct = {a:pow(a,3) for a in range(1,100)}
print(dct)

# Llaves: Los 1000 primeros números; Valores: La raíz cuadrada de la llave
dct = {a:pow(a,1/3) for a in range(1,1000)}
print(dct)
````
